```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0, 10, 30) + np.random.normal(size=30)
x
```

```python
y = x**2
```

```python
df = pd.concat([
    pd.Series(x),
    pd.Series(y),
], axis=1)

df.columns=["x", "y"]

df.head()
```

```python
sns.lmplot(data=df, x = "x", y="y")
plt.title("Quadratic relationship")
```

```python
df["x"] = df["x"]**2
```

```python
sns.lmplot(data=df, x = "x", y="y")
plt.title("Linear relationship with quadratic feature")
plt.xlabel("x squared")
```
